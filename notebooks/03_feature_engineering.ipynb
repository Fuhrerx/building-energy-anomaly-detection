# Feature Engineering

## What we did so far

We Preprocessed our data and made changes to the timestamp column to a much more standardized format for easier navigation and visualisation in the future. We also addressed outlier issues by capping the dataset to 1st percentile and 99th percentile removing extreme ends that may skew our predictions and anomalies.

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
```

```python
df = pd.read_csv(r"C:\Users\Akshay\Desktop\building-energy-anomaly-detection\data\meters\whole\eda.csv")
```

```python
# Select numeric columns for modeling (exclude timestamp)
numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()

# Remove timestamp from numeric_cols if present
if 'timestamp' in numeric_cols:
    numeric_cols.remove('timestamp')

print(f"Number of numeric features: {len(numeric_cols)}")

# Check if we have any numeric columns
if len(numeric_cols) == 0:
    raise ValueError("No numeric columns found in the dataset!")

# Prepare feature matrix with NaN handling
X = df[numeric_cols].fillna(0).values

# Handle any remaining NaN values
X = np.nan_to_num(X, nan=0.0, posinf=0.0, neginf=0.0)

print(f"Feature matrix shape: {X.shape}")
```

## Outcomes of Feature Enigineering

We have did the following steps to engineer our features :

- Selected pure numeric columns to transform into features for our model
- Removed timestamp column from newly created numeric_cols 
- Checked the number of numeric columns we created
- Validating the number of numeric columns
- Filled NaN and NULL values
